## 1.2 Scrapping Function

    - the input of the function will be a list of urls, and the output csv files with the principals features of the house:
        -decription
        -location
        -number of rooms
        -number of bathrooms
        -square meters surface buidl
        -square meters surface terrain.

```python
def scra_engels(name ,weblist):
    
    csv_file = open(f'{name}.csv', 'w')#csv that save

    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['description','location','attributs','price'])#headers of the csv
    
    for url in weblist:
    
        source = requests.get(url).text

        soup = BeautifulSoup(source, 'lxml')


        for prop in soup.find_all('div', class_='ev-teaser-content'):
            title = str(prop.find('div', class_='ev-teaser-title').text)
            print(title)
            subtitle = prop.find('div', class_='ev-teaser-subtitle').text
            print(subtitle)
            attributes = prop.find('div', class_='ev-teaser-attributes').text
            print(attributes)
            price = prop.find('div', class_='ev-value').text
            print(price)
            print('++++++++++++++++++++++++++++++++++++++++++++++++++')
            csv_writer.writerow([title,subtitle,attributes,price])

    csv_file.close()
```
